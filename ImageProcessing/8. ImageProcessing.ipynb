```python
# Loading Libraries
import numpy as np 
import matplotlib.pylab as plt

%matplotlib inline


img1 = plt.imread("/Users/pavitragajanana/development/5. InternalFiles/Image1.jpg")
img1.shape

def plti(img, h=10, **kwargs):
    """
    Helper function to plot an image.
    """
    y = img.shape[0]
    x = img.shape[1]
    w = (y/x) * h
    plt.figure(figsize=(w,h))
    plt.imshow(img, interpolation="none", **kwargs)
    plt.axis('off')

plti(img1,10)
```
